## Let's define some function we gonna use in the crawling process

```python
def initDriver():
    """
    Initialize the Chrome driver with options to disable notifications.
    This effectively creates a new browser window.
    """
    options = webdriver.ChromeOptions()
    options.add_argument("--disable-notifications")
    return webdriver.Chrome(options=options)

def fbLogin(driver, email, password):
    """
    Log in to Facebook using the provided email and password after 
    the browser window is opened.
    """
    driver.get("https://facebook.com")

    # Login process
    userEmail = driver.find_element(by="id", value="email")
    userEmail.send_keys(email)
    userPass = driver.find_element(by="id", value="pass")
    userPass.send_keys(password)
    userPass.send_keys(Keys.ENTER)

def readData(fileName):
    """ 
    Read data from a given file, mainly used to read a list of post IDs.
    """
    with open(fileName, 'r', encoding='utf-8') as f:
        data = [line.strip() for line in f]
    return data

def writeFileTxt(fileName, content):
    """ 
    Write the given content to a file, mainly used to save a list of post IDs.
    """
    with open(fileName, 'a', encoding='utf-8') as f1:
        f1.write(content + '\n')

def getPostID(driver, page_id, all_posts_set, filePath='posts.csv'):
    """ 
    Get all post IDs on the current page and scroll down to load more posts.
    """
    sleep(5)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    elements = driver.find_elements(By.XPATH, f'//a[contains(@href, "https://www.facebook.com/{page_id}/posts/")]')
    
    for element in elements:
        postId = element.get_attribute("href").split('posts/')[1].split('?')[0]
        if postId not in all_posts_set:
            print(postId)
            writeFileTxt(filePath, postId)
            all_posts_set.add(postId)

def getnumPost(driver, page_id, amount, filePath='posts.csv'):
    """ 
    Use the getPostID function to get post IDs until reaching the desired amount.
    """
    all_posts_set = set(readData(filePath))
    driver.get("https://facebook.com/" + page_id)
    while len(all_posts_set) < amount:
        getPostID(driver, page_id, all_posts_set, filePath)
```

```python
post_IDs = readData('posts.csv') # List of post IDs
print(len(post_IDs)) # Got >= 100 IDs due to over-scrolling
```

```python
post_urls = [f"https://facebook.com/{id}" for id in post_IDs]
options = {
    "comments": 10,
    "reactions" : True,
    "allow_extra_requests": True,
}
posts = []

# Try to get as many post as possible for one account
for post in get_posts(
    post_urls=post_urls, 
    credentials=(EMAIL, PASSWORD),
    options=options,
):
    posts.append(post)
    sleep(20)
```
